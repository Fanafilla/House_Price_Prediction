## Seperate independent & dependent variable into x and y

```python
x = x_df.drop('SalePrice', axis=1)
y = x_df['SalePrice']
```

```python
x.head()
```

```python
x.columns
```

```python
y.head()
```

### Splitting the data set into test and train sets

```python
from sklearn.model_selection import train_test_split, KFold , StratifiedKFold , cross_val_score, GridSearchCV
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.20, random_state=42)
```

### Create Linear Regression Model

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
model = LinearRegression()
model.fit(xtrain, ytrain)
y_pred = model.predict(xtest)
y_pred
```

```python
model.score(xtest, ytest)
```

```python
# Cross validation to find 'validat' score apply multiple samples , automaticaly does Kfold stratifying

# KFold Cross validation
kfold = KFold(n_splits=5)
model_cv = cross_val_score(model, x, y , cv=kfold)
model_cv
```

```python
cross_validation_result = model_cv.mean()
cross_validation_result
```

## Support Vector Machines Model

```python
from sklearn.svm import SVR
svr = SVR(C=100000)
svr.fit(xtrain, ytrain)
svr.predict(xtest)
```

```python
svr.score(xtest, ytest)
```

```python
# KFold Cross validation

kfold = KFold(n_splits=5)
model_cv = cross_val_score(svr, x, y , cv=kfold)
model_cv.mean()
```

## Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(xtrain, ytrain)
rfr.predict(xtest)
```

```python
rfr.score(xtest, ytest)
```

```python
# KFold Cross validation

kfold = KFold(n_splits=5)
model_cv = cross_val_score(rfr, x, y , cv=kfold)
model_cv.mean()
```

## XGBoost Regressor Model

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(n_estimators=1000, learning_rate=0.01)
xgb.fit(xtrain, ytrain)
xgb.predict(xtest)
```

```python
xgb.score(xtest, ytest)
```

```python
# KFold Cross validation

kfold = KFold(n_splits=5)
model_cv = cross_val_score(xgb, x, y , cv=kfold)
model_cv.mean()
```

## Polynomial Regression (Degree=2)

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
xtrain_2d = poly_reg.fit_transform(xtrain)
xtest_2d = poly_reg.transform(xtest)

lin_reg = LinearRegression()
lin_reg.fit(xtrain_2d, ytrain)
lin_reg.predict(xtest_2d)
```

```python
lin_reg.score(xtest_2d, ytest)
```

```python
# KFold Cross validation

kfold = KFold(n_splits=5)
model_cv = cross_val_score(lin_reg, x, y , cv=kfold)
model_cv.mean()
```
